Step 1 - Create dataset

```python
import numpy as np

# 1.1 Input generation
X = np.linspace(0, 10, 50).reshape(-1, 1)

# 1.2 Target generation
noise = np.random.normal(0, 0.2, size=(50, 1))
y = np.log(X + 1) + noise

# why this target cannot be fit well by a straight line
# logarithmic shape grows fast initially and slows later. 
# A straight line has only one slope and cannot match both regions simultaneously
```

Step 2 - Decide the Model Shape

- How many intermediate values (hidden units) do you want?
    * choose 3
- Why more than 1?
- Why not too many?
    * you are fixing the shape of the model, not tuning numbers.

Hidden units allow multiple “pieces” of behavior. Too many units increase
complexity and instability; too few cannot bend enough.

Step 3 - Initialize model param

```python
W1 = np.random.uniform(-1,1, size=(1,3))
b1 = np.zeros((1,3))

W2 = np.random.uniform(-1,1, size=(3,1))
b2 = np.zeros((1,1))

# what each parameter controls?
# which ones affect shape vs position?
# Weights affect shape (slope and bending). Biases shift the curve up or down.
```

Step 4 - Define an Activation Function

```python
def activation(z):
    return np.maximum(0, z)

def activation_slope(z):
    return (z > 0).astype(float)
```

Step 5 - Forward Pass

```python
z1 = X @ W1 + b1
h = activation(z1)
y_hat = h @ W2 + b2
```

Step 6 - Define Error and Loss

```python
error = y_hat - y
loss = np.mean(error **2)
```

Step 7 - Why We Need Slopes?

- To reduce loss, we must change parameters.
- To know how to change them, we need to know:
- if a small change increases or decreases loss.

Step 8 - Backward Flow

```python
# loss to prediction
dL_dy = 2*error /len(X)

# gradient of o.p layer
dL_dW2 = h.T @ dL_dy
dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

# why these depend on hidden values?
# becz during forward prop ans comes through these layers, so during back prop we adjust
# weights of hidden layer to reduce error

# flow into hidden layer
dL_dh = dL_dy @ W2.T

# apply activation slope
dL_dz1 = dL_dh * activation_slope(z1)
# how activation controls error flow
# its get multiplied by the error

# gradient of 1st layer
dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
```

Step 9 - Parameter Update Rule

```python
learning_rate =0.01
W1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1
W2 -= learning_rate * dL_dW2
b2 -= learning_rate * dL_db2
```

Step 10 - Training Loop

```python
for epoch in range(1000):
    # forward pass
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    # loss
    error = y_hat - y
    loss = np.mean(error **2)
    
    # backward pass
    dL_dy =2 * error /len(X)
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)
    
    dL_dh = dL_dy @ W2.T
    dL_dz1 = dL_dh * activation_slope(z1)
    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
    
    # update params
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2
    
    if epoch %100 ==0:
        print(f"Epoch {epoch}: {loss}")
    
```
