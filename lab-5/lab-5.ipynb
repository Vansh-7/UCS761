```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)

def generate_data():
    length = np.random.normal(0.523992, 0.120093, 100)
    height = np.random.normal(0.139516, 0.041827, 100)
    diameter = np.random.normal(0.139516, 0.041827, 100)
    whole_w = np.random.normal(0.828742, 0.490389, 100)
    shucked_w = np.random.normal(0.359367, 0.221963, 100)
    viscera_W = np.random.normal(0.180594, 0.109614, 100)
    shell_w = np.random.normal(0.238831, 0.139203, 100)
    
    y = 0.5 * shell_w + 0.3 * whole_w + 0.2 * diameter
        
    df = pd.DataFrame({'length' : length, 'height': height, 'diameter': diameter, 'whole_weight': whole_w, 'shucked_weight': shucked_w, 'viscera_weight': viscera_W, 'shell_weight': shell_w, 'target': y})
    
    return df
    
df = generate_data()
df.head()
```

```python
input_var = ['shell_weight', 'whole_weight', 'diameter'] 

X = df[input_var]
y = df['target']
```

```python
w = np.zeros(3)
b = 0.0

def predict(X, w, b):
    y_cap = X.dot(w) + b
    return y_cap

def mean_sq_err(y, y_cap):
    loss = np.mean((y_cap - y)**2)
    return loss

def compute_gradients(X, y, y_pred):
    n = len(y)
    
    dw = (2/n) * X.T.dot(y_pred-y)
    db = (2/n) * (y_pred-y).sum()
    
    return dw, db

def update_param(w, b, dw, db, lr):
    w = w - lr*dw
    b = b - lr*db

    return w, b
```

```python
lr = 0.0001
epochs = 1000

for e in range(epochs):
    y_cap = predict(X, w, b)
    loss = mean_sq_err(y, y_cap)
    dw, db = compute_gradients(X, y, y_cap)
    w, b = update_param(w, b, dw, db, lr)
    
    if e % 100 == 0:
        print(f'Epoch {e}, Loss: {loss}')
```

```python
new_sample = np.array([0.2, 0.6, 0.4])
predicted_age = new_sample.dot(w) + b
print(predicted_age)
```
